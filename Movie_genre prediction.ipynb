```python
#importing essential libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
#loading dataset
df = pd.read_csv("train_data.txt",sep=":::", engine="python", header=None,names=["id","title","genre","description"])
df.head(6)
```

```python
#checking if any NaN values are present or notprint(df.isnull().sum())
print(df.isnull().sum())
```

```python
#Converting text to lowercase and removing extra whitespace
import re
def text(t):
    t=t.lower()
    t= re.sub(r'\s+', ' ', t).strip()
    return t
df['description'] = df['description'].apply(text)
```

```python
#selecting columns
df=df[["description","genre"]]
```

```python
#Determinig independent and dependent variables
x=df["description"]
y=df["genre"]
```

```python
#splitting the data into train and test 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=30)
```

```python
#Text preprocessing
encoder=TfidfVectorizer(stop_words="english",max_features=15000,ngram_range=(1,2))
x_train_vect=encoder.fit_transform(x_train)
x_test_vect=encoder.transform(x_test)
```

```python
#Selecting the training model
NB=MultinomialNB()
```

```python
#Training the model
NB=NB.fit(x_train_vect,y_train)
NB
```

```python
y_predict=NB.predict(x_test_vect)
```

```python
#Checking the accuracy
accuracy=accuracy_score(y_test,y_predict)
print("Accuracy score is:",accuracy)
```

```python
print(classification_report(y_test,y_predict,zero_division=0))
```
